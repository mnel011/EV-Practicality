# First attempt at webscraping

```python
import requests as rqt
from bs4 import BeautifulSoup as soup
import re
```

```python
url = 'https://ev-database.org/cheatsheet/acceleration-electric-car'
```

```python
response = rqt.get(url)
```

```python
html = response.content
```

```python
html_bs = soup(html, 'html.parser')
```

```python
print(html_bs.prettify())
```

```python
tag1 = html_bs.find_all('span', class_='bar')
print(tag1[0])
```

```python
accel = []
for i in range(0,len(tag1)):
    #print(str(tag1[i]))
    accel.append(list(re.findall('\d+.\d', str(tag1[i])))[0])
print(accel)    
```

```python
tag2 = html_bs.find_all('a', target='_blank')
print(tag2[0].text)
```

```python
car = []
for i in range(0,len(tag2)):
    car.append(str(tag2[i].text))
print(car)    
```

```python
dict1 = {}
for x in range(0,len(car)):
    dict1[car[x]] = accel[x]

print(dict1)    
```
